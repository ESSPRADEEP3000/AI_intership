```python
#task-1 ->Read the given three CSV files to different pandas dataframes.
import pandas as pd
customers_df=pd.read_csv('/content/drive/MyDrive/internship/Customers.csv')
transaction_df=pd.read_csv('/content/drive/MyDrive/internship/Transaction.csv')
studies_df=pd.read_csv('/content/drive/MyDrive/internship/Studies.csv')
```

```python
#task-2 ->merging the customers and transactions dataframes into one dataframe
merged_df=pd.merge(customers_df,transaction_df,on="customer_id")
merged_df
```

```python
#task-3 -> Duration in days between start date and end date
merged_df['start_date']= pd.to_datetime(merged_df['start_date'],format="%d-%m-%Y")
#some end dates are causing out of bound errors to deal with we find such dates first
problamatic_dates=merged_df[merged_df['end_date']=='31-12-9999']
# print(problamatic_dates)
#now we replace such dates with a text NaT (or) replace with the final date
merged_df.loc[merged_df['end_date']=='31-12-9999','end_date']=pd.NaT
merged_df['end_date']= pd.to_datetime(merged_df['end_date'],format="%d-%m-%Y")
# Timedelta object representing the time difference.
# .dt.days accesses the days attribute of the Timedelta object, returning the duration in days.
merged_df['duration']=(merged_df['end_date']-merged_df['start_date']).dt.days
merged_df
```

```python
#task-4 -> Drop the duplicate rows in the merged dataframe
merged_df.drop_duplicates(inplace=True)
merged_df
```

```python
#task-5 ->Drop rows with missing values in the merged dataframe
merged_df.dropna(inplace=True)
merged_df
```

```python
#task-6 ->Calculating the average duration of each customer.
merged_df['average_duration']=merged_df.groupby('customer_id')['duration'].mean().round(2)
merged_df
```

```python
#task-7. Display the unique transaction types.
uni_txn_types=merged_df['txn_type'].unique()
print(uni_txn_types)
```

```python
#TASK-8 -> count of each transaction type with respect to each continent
continent_df=pd.read_csv('/content/drive/MyDrive/internship/Continent.csv')
merged_df=pd.merge(customers_df,transaction_df,on="customer_id")
merged_df2=pd.merge(merged_df,continent_df,on="region_id")
count_txn_type_continent=merged_df2.groupby(['txn_type','region_name']).size()
print(count_txn_type_continent)
```
