```python
#task-6 ->Calculating the average duration of each customer.
import pandas as pd
customers_df=pd.read_csv('/content/drive/MyDrive/internship/Customers.csv')
transaction_df=pd.read_csv('/content/drive/MyDrive/internship/Transaction.csv')
continent_df=pd.read_csv('/content/drive/MyDrive/internship/Continent.csv')
merged_df=pd.merge(customers_df,transaction_df,on="customer_id")
merged_df2=pd.merge(merged_df,continent_df,on="region_id")
merged_df2['start_date']= pd.to_datetime(merged_df2['start_date'],format="%d-%m-%Y")
#some end dates are causing out of bound errors to deal with we find such dates first
problamatic_dates=merged_df2[merged_df2['end_date']=='31-12-9999']
# print(problamatic_dates)
#now we replace such dates with a text NaT (or) replace with the final date
merged_df2.loc[merged_df2['end_date']=='31-12-9999','end_date']=pd.NaT
merged_df2['end_date']= pd.to_datetime(merged_df2['end_date'],format="%d-%m-%Y")
# Timedelta object representing the time difference.
# .dt.days accesses the days attribute of the Timedelta object, returning the duration in days.
merged_df2['duration']=(merged_df2['end_date']-merged_df2['start_date']).dt.days
merged_df2
```

```python
merged_df2.drop_duplicates(inplace=True)
merged_df2.dropna(inplace=True)
merged_df2
```

```python
merged_df2['average_duration']=merged_df2.groupby('customer_id')['duration'].mean().round(2)
merged_df2
```
